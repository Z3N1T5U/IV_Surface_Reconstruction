```python
import numpy as np
from scipy.interpolate import interp1d

print("Applying Quant-Standard Variance Interpolation...")

def fix_row_variance(strikes, iv_values):
    mask = ~np.isnan(iv_values)
    known_strikes = strikes[mask]
    known_iv = iv_values[mask]

    if len(known_strikes) < 2:
        return iv_values

    sort_idx = np.argsort(known_strikes)
    known_strikes = known_strikes[sort_idx]
    known_iv = known_iv[sort_idx]

    # --- THE MAGIC FIX: Convert to Variance Space ---
    known_variance = known_iv ** 2

    # Interpolate the Variance, not the Volatility
    interp_func = interp1d(
        known_strikes, known_variance, kind="linear",
        fill_value="extrapolate", bounds_error=False
    )

    # Predict the Variance for all strikes
    predicted_variance = interp_func(strikes)

    # Safety feature: Prevent math errors from tiny negative extrapolations
    predicted_variance = np.maximum(predicted_variance, 1e-10)

    # --- Convert back to Volatility Space ---
    return np.sqrt(predicted_variance)

# Create a new dataset for these predictions
df_variance = df.copy()

print("Calculating the new U-shaped curves. Please wait...")
for index, row in df.iterrows():
    row_numbers = row[option_cols].values.astype(float)
    df_variance.loc[index, option_cols] = fix_row_variance(strikes, row_numbers)

# Save the file
output_name = "submission_variance_space.csv"
df_variance.to_csv(output_name, index=False)
print(f"✅ Finished! File saved as {output_name}")
```

```python
import pandas as pd
import numpy as np
from scipy.interpolate import interp1d
import time
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error

print("🚀 INITIATING ULTIMATE MODEL EVALUATION SUITE 🚀\n")

# --- ALGORITHM DEFINITIONS ---
def old_linear(strikes, iv_values):
    mask = ~np.isnan(iv_values)
    if len(strikes[mask]) < 2: return iv_values
    func = interp1d(strikes[mask], iv_values[mask], kind="linear", fill_value="extrapolate", bounds_error=False)
    return func(strikes)

def new_variance_clamped(strikes, iv_values):
    mask = ~np.isnan(iv_values)
    known_strikes = strikes[mask]
    known_iv = iv_values[mask]
    if len(known_strikes) < 2: return iv_values

    sort_idx = np.argsort(known_strikes)
    known_strikes = known_strikes[sort_idx]
    known_iv = known_iv[sort_idx]

    known_variance = known_iv ** 2
    interp_func = interp1d(
        known_strikes, known_variance, kind="linear",
        fill_value=(known_variance[0], known_variance[-1]), bounds_error=False
    )
    predicted_variance = interp_func(strikes)
    predicted_variance = np.maximum(predicted_variance, 1e-10)
    return np.sqrt(predicted_variance)

# --- TEST 1: EXECUTION TIMING ---
print("⏱️ TEST 1: BENCHMARKING SPEED...")
test_rows = 500 # Test on a sample to get average speed

start_time = time.time()
for _, row in df.head(test_rows).iterrows():
    old_linear(strikes, row[option_cols].values.astype(float))
old_time = time.time() - start_time

start_time = time.time()
for _, row in df.head(test_rows).iterrows():
    new_variance_clamped(strikes, row[option_cols].values.astype(float))
new_time = time.time() - start_time

print(f"Old Method Time (per 500 rows): {old_time:.4f} seconds")
print(f"New Method Time (per 500 rows): {new_time:.4f} seconds")
if new_time < old_time * 1.5:
    print("✅ PASS: New method is highly optimized and will not time out on Kaggle.\n")


# --- TEST 2: THE KAGGLE SIMULATOR (CROSS-VALIDATION) ---
print("🎯 TEST 2: LOCAL KAGGLE LEADERBOARD SIMULATOR (MSE calculation)...")
# We will deliberately hide 20% of KNOWN values, predict them, and check our accuracy.

np.random.seed(42) # For reproducibility
true_values_list = []
old_preds_list = []
new_preds_list = []

# Test on 100 random rows to simulate the leaderboard
sample_df = df.dropna(subset=option_cols, how='all').sample(100, random_state=42)

for index, row in sample_df.iterrows():
    row_iv = row[option_cols].values.astype(float)
    mask = ~np.isnan(row_iv)
    valid_indices = np.where(mask)[0]

    if len(valid_indices) > 5:
        # Randomly select 20% of known values to hide
        num_to_hide = max(1, int(len(valid_indices) * 0.2))
        hide_indices = np.random.choice(valid_indices, num_to_hide, replace=False)

        # Create the damaged row
        damaged_iv = row_iv.copy()
        damaged_iv[hide_indices] = np.nan

        # Predict with both models
        old_pred_row = old_linear(strikes, damaged_iv)
        new_pred_row = new_variance_clamped(strikes, damaged_iv)

        # Save the results for the hidden spots
        for idx in hide_indices:
            true_values_list.append(row_iv[idx])
            old_preds_list.append(old_pred_row[idx])
            new_preds_list.append(new_pred_row[idx])

# Calculate Kaggle's favorite metric: Mean Squared Error
old_mse = mean_squared_error(true_values_list, old_preds_list)
new_mse = mean_squared_error(true_values_list, new_preds_list)

print(f"Old Baseline Simulated Kaggle MSE: {old_mse:.8f}")
print(f"New Variance Simulated Kaggle MSE: {new_mse:.8f}")

improvement = old_mse - new_mse
if improvement > 0:
    print(f"✅ PASS: New method is BETTER by {improvement:.8f} MSE!")
else:
    print(f"⚠️ WARNING: Old method scored better on this specific random sample.")


# --- TEST 3: VISUALIZING THE VOLATILITY SMILE ---
print("\n📊 TEST 3: VISUALIZING THE DIFFERENCE...")
# Find a row with a big gap in the middle to show the convexity
missing_counts = df[option_cols].isna().sum(axis=1)
best_row_idx = missing_counts.idxmax()

raw_row = df.loc[best_row_idx, option_cols].values.astype(float)
old_curve = old_linear(strikes, raw_row)
new_curve = new_variance_clamped(strikes, raw_row)

plt.figure(figsize=(12, 6))
plt.plot(strikes, old_curve, label='Old: Rigid Linear Extrapolation', linestyle='--', color='red', marker='x')
plt.plot(strikes, new_curve, label='New: Variance Space Clamped', color='blue', marker='o', linewidth=2)
# Highlight original known points
mask = ~np.isnan(raw_row)
plt.scatter(strikes[mask], raw_row[mask], color='green', s=100, zorder=5, label='Original Known Data')

plt.title(f"Volatility Smile Comparison (Row {best_row_idx})")
plt.xlabel("Strike Price")
plt.ylabel("Implied Volatility")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

print("\n==========================================")
print("FINAL ASSESSMENT COMPLETE.")
```
